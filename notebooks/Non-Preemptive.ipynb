```python
def calculate_sjf_times(burst_times):
    num_processes = len(burst_times)

    # Initialize lists to store times
    waiting_times = [0] * num_processes

    turnaround_times = [0] * num_processes
    completion_times = [0] * num_processes

    # Create a list of process indices and sort based on burst time
    sorted_indices = sorted(range(num_processes), key=lambda i: burst_times[i])
    
    current_time = 0
    
    for i in range(num_processes):
        index = sorted_indices[i]

        # Calculate completion time
        completion_times[index] = current_time + burst_times[index]

        # Calculate turnaround time
        turnaround_times[index] = completion_times[index] # Arrival time is 0

        # Calculate waiting time
        waiting_times[index] = turnaround_times[index] - burst_times[index]

        # Update current time
        current_time = completion_times[index]
    return completion_times, turnaround_times, waiting_times
    
def print_table2(arrival_times, burst_times):
    print(f"{'Process':<10} {'AT':<10} {'BT':<10}")
    print("-" * 30)
    
    for i in range(len(burst_times)):
        print(f"Process {i+1:<2} {arrival_times[i]:<10} {burst_times[i]:<10}")

def print_table(arrival_times, burst_times, completion_times, turnaround_times, waiting_times):
    print(f"{'Process':<10} {'AT':<10} {'BT':<10} {'CT':<10} {'TAT':<10} {'WT':<10}")
    print("-" * 60)
    
    for i in range(len(burst_times)):
        print(f"Process {i+1:<2} {arrival_times[i]:<10} {burst_times[i]:<10} {completion_times[i]:<10} {turnaround_times[i]:<10} {waiting_times[i]:<10}")

def print_gantt_chart(arrival_times, burst_times, completion_times):
    num_processes = len(burst_times)

    # Sort processes by burst times for SJF execution order
    sorted_indices = sorted(range(num_processes), key=lambda i: burst_times[i])

    gantt_chart = "Gantt Chart:\n"
    gantt_line = "0"  # Start time marker
    time_marker_line = ""

    current_time = 0

    for idx in sorted_indices:
        # Add idle time if CPU is idle
        if current_time < arrival_times[idx]:
            gantt_line += f"->IDLE->"
            time_marker_line += f"{current_time:<6}"  # Mark idle start

            # Jump to the process's arrival time
            current_time = arrival_times[idx]

        # Add process block to Gantt chart
        gantt_line += f"P{idx + 1}->"
        time_marker_line += f"{current_time:<6}->"  # Mark process start

        # Update current time after execution of the process
        current_time += burst_times[idx]

    # Add the final time marker
    gantt_line += f"{current_time}"  # Final completion time
    time_marker_line += f"{current_time:<6}"  # Final time marker

    print(gantt_chart)
    print(gantt_line)
    print(time_marker_line)


def main():
    # Input number of processes
    num_processes = int(input("Enter the number of processes: "))
    
    # Input arrival times
    arrival_times = []
    print("Enter the arrival times for each process:")
    for i in range(num_processes):
        arrival_time = int(input(f"Enter the Arrival time for process P{i+1}: "))
        arrival_times.append(arrival_time)
    
    # Input burst times
    burst_times = []
    print("Enter the burst times for each process:")
    for i in range(num_processes):
        burst_time = int(input(f"Enter the Burst time for process P{i+1}: "))
        burst_times.append(burst_time)

    # Calculate times using SJF
    completion_times, turnaround_times, waiting_times = calculate_sjf_times(burst_times)
    
    # Calculate average waiting time and average turnaround time
    average_waiting_time = sum(waiting_times) / num_processes
    average_turnaround_time = sum(turnaround_times) / num_processes
    
    #Print the Inputs
    print("\n")
    print("Entered inputs are :")
    print_table2(arrival_times, burst_times)
    print("\n")
    
    # Print Gantt Chart
    print_gantt_chart(arrival_times, burst_times, completion_times)
    print("\n")

    # Print the results in table format
    print_table(arrival_times, burst_times, completion_times, turnaround_times, waiting_times)
    
    # Print average waiting time and average turnaround time
    print(f"\nAverage Turnaround Time: {average_turnaround_time:.2f}")
    print(f"Average Waiting Time: {average_waiting_time:.2f}")
    
if __name__ == "__main__":
    main()
```

### SJF

```python
def sjf_time(arrival_time,burst_time):
    n=len(arrival_time)
    completion_time=[0]*n
    turnaround_time=[0]*n
    waiting_time=[0]*n
    start_time=0
    sorted_idx=sorted(range(n),key=lambda x:burst_time[x])
    for i in range(n):
        idx=sorted_idx[i]
        completion_time[idx]=start_time+burst_time[idx]
        turnaround_time[idx]=completion_time[idx]-arrival_time[idx]
        waiting_time[idx]=turnaround_time[idx]-burst_time[idx]
        start_time=completion_time[idx]
    return completion_time,turnaround_time,waiting_time

def print_table(arrival_time,burst_time):
    print(f"{'Process':<10} {'AT':<10} {'BT':<10}")
    print('-'*30)
    for i in range(len(arrival_time)):
        print(f"P{i:<10} {arrival_time[i]:<10} {burst_time[i]:<10}")

def print_table2(arrival_time,burst_time,completion_time,turnaround_time,waiting_time):
    print(f"{'Process':<10} {'AT':<10} {'BT':<10} {'CT':<10} {'TAT':<10} {'WT':<10}")
    print('-'*60)
    for i in range(len(arrival_time)):
        print(f"P{i:<10} {arrival_time[i]:<10} {burst_time[i]:<10} {completion_time[i]:<10} {turnaround_time[i]:<10} {waiting_time[i]:<10}")

def gantt_chart(arrival_time,burst_time,completion_time,turnaround_time,waiting_time):
    start_time=0
    sorted_idx=sorted(range(len(arrival_time)),key=lambda x:burst_time[x])
    gantt_line='0'
    time_line=""
    for idx in sorted_idx:
        if start_time<arrival_time[idx]:
            gantt_line+='->Idle'
            time_line+=f'{start_time}'
            start_time=arrival_time[idx]
        gantt_line+=f'->P{idx}'
        time_line+=f'{start_time}'
        start_time+=burst_time[idx]
    gantt_line+=f'{start_time}'
    time_line+=f'{start_time}'
    print(gantt_line)
    print(time_line)
n=int(input("Enter no. of processes:"))
arrival_time=[]
burst_time=[]
for i in range(n):
    tm=int(input(f"Enter arrival time for process P{i}"))
    arrival_time.append(tm)
for i in range(n):
    tm=int(input(f"Enter burst time for process P{i}"))
    burst_time.append(tm)
print(arrival_time)
print(burst_time)
print_table(arrival_time=arrival_time,burst_time=burst_time)
completion_time,turnaround_time,waiting_time=sjf_time(arrival_time,burst_time)
print_table2(arrival_time,burst_time,completion_time,turnaround_time,waiting_time)
gantt_chart(arrival_time,burst_time,completion_time,turnaround_time,waiting_time)
```

```python
def calc_priority(arrival_time,burst_time,priorities):
    n=len(arrival_time)
    completion_time=[0]*n
    turnaround_time=[0]*n
    waiting_time=[0]*n
    sorted_idx=sorted(range(n),key=lambda x:priorities[x])
    current=0
    for i in range(n):
        idx=sorted_idx[i]
        completion_time[idx]=current+burst_time[idx]
        turnaround_time[idx]=completion_time[idx]-arrival_time[idx]
        waiting_time[idx]=turnaround_time[idx]-burst_time[idx]
        current=completion_time[idx]
    return completion_time,turnaround_time,waiting_time

def print_table(arrival_time,burst_time,priorities):
    print(f"{'Process':<10} {'AT':<10} {'BT':<10} {'Priority':<10}")
    print('-'*30)
    for i in range(len(arrival_time)):
        print(f"P{i:<10} {arrival_time[i]:<10} {burst_time[i]:<10} {priorities[i]:<10}")
def print_table2(arrival_time,burst_time,priorities,completion_time,turnaround_time,waiting_time):
    print(f"{'Process':<10} {'AT':<10} {'BT':<10} {'Prior':<10}{'CT':<10} {'TAT':<10} {'WT':<10}")
    print('-'*60)
    for i in range(len(arrival_time)):
        print(f"P{i:<10} {arrival_time[i]:<10} {burst_time[i]:<10} {priorities[i]:<10} {completion_time[i]:<10} {turnaround_time[i]:<10} {waiting_time[i]:<10}")
        
n=int(input("Enter no. of processes:"))
arrival_time=[]
burst_time=[]
priorities=[]
for i in range(n):
    tm=int(input(f"Enter arrival time for process P{i}"))
    arrival_time.append(tm)
for i in range(n):
    tm=int(input(f"Enter burst time for process P{i}"))
    burst_time.append(tm)
for i in range(n):
    tm=int(input(f"Enter priority for process P{i}"))
    priorities.append(tm)
print(arrival_time)
print(burst_time)
print_table(arrival_time=arrival_time,burst_time=burst_time,priorities=priorities)
completion_time,turnaround_time,waiting_time=calc_priority(arrival_time,burst_time,priorities)
print_table2(arrival_time,burst_time,priorities,completion_time,turnaround_time,waiting_time)
```
